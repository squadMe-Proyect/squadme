## Importamos pandas y subimos el archivo csv/json

```python
import pandas as pd
players = pd.read_csv("./sample_data/players.csv")
players
```

## 1 - Eliminar espacios al final de cada línea: Hacemos por ejemplo en la columna name

```python
players['name'] = players['name'].str.rstrip()
players
```

## 2 - Sustituir carácteres: Sustituyo los valores en position por sus tres primeros carácteres

```python
players['position'] = players[['position']].replace(['Delantero','Centrocampista','Defensa','Portero'],['Del','Cen','Def','Por'])
players
```

## 3 - Sustituir espacios en blanco o falta de información: Sustituyo el valor NaN de la columna picture por "No tiene foto"



```python
players['picture'] = players.picture.fillna('No tiene foto')
players
```

## 4 - Realizar búsquedas en el dataframe: Jugadores que tengan más de 5 o más asistencias

```python
plys_assists = players.loc[players['assists'] >= 5, ['name','surname']]
plys_assists
```

## 5 - Crear nuevas columnas en base a cálculos realizados: Sumo goles y asistencias y calculo las aportaciones de gol (goal_contributions)

```python
goal_contributions = players.goals + players.assists
df_goal_contributions = pd.DataFrame({'goal_contributions': goal_contributions})
players = players.join(df_goal_contributions)
players
```

## Crear nuevo fichero de salida

```python
players.to_csv('players-modified.csv', index=False)
```
